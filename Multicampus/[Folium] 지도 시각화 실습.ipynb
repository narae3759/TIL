# 공공데이터 포털 : 서울특별시 버스정보
* serviceKey: 내가 활용 신청을 했을 때 '일반 인증키'를 복사해서 붙인다.

```python
import requests

url = 'http://ws.bus.go.kr/api/rest/arrive/getArrInfoByRouteAll'
params ={'serviceKey' : '****************', 
         'busRouteId' : '*****************' }

response = requests.get(url, params=params)
# response.text
# print(response.content)
```

```python
## Decoding 
xml_doc = response.content.decode('utf-8')
type(xml_doc)
```

```python
# xml -> JSON
import xmltodict 
json_doc = xmltodict.parse(xml_doc)
type(json_doc)
```

```python
json_doc['ServiceResult']['msgBody']['itemList'][0]
```

```python
# 모든 노선의 정류장 ID를 정류장 명칭으로 바꾸기 (리스트로) - 강사님
df_station = pd.read_csv('data/서울시버스정류소좌표데이터(2022.08.24).csv', encoding='CP949')
df_station = df_station.set_index('NODE_ID')
st_id_lst = json_doc['ServiceResult']['msgBody']['itemList']

for st in st_id_lst: 
    st_id = int(st['stId'])
    st_name = df_station.loc[st_id, '정류소명']
    longitude = df_station.loc[st_id, '좌표Y']
    latitude = df_station.loc[st_id, '좌표X']
    
    print([st_id, st_name, latitude, longitude])
```

```python
# 모든 노선의 정류장 ID를 정류장 명칭으로 바꾸기 (리스트로)
df_station = pd.read_csv('data/서울시버스정류소좌표데이터(2022.08.24).csv', encoding='CP949')
df_station = df_station.drop(labels='ARS-ID', axis=1)
st_id_lst = json_doc['ServiceResult']['msgBody']['itemList']

for st in st_id_lst: 
    print(list(df_station[df_station.NODE_ID == int(st['stId'])].iloc[0]))
```

```python
# 모든 노선의 정류장 ID를 정류장 명칭으로 바꾸기 (리스트로)
df_station = pd.read_csv('data/서울시버스정류소좌표데이터(2022.08.24).csv', encoding='CP949')
df_station = df_station.set_index('NODE_ID')

st_id_lst = json_doc['ServiceResult']['msgBody']['itemList']

st_lst = []
for st in st_id_lst: 
    st_id = int(st['stId'])
    lst = [st_id]
    lst.extend(list(df_station[['정류소명','좌표X','좌표Y']].loc[st_id]))
    
    st_lst.append(lst)

st_lst
```

```python
import folium

df_st = pd.DataFrame(st_lst,columns=['ID','정류소명','경도','위도'])
df_st = df_st.set_index('ID')
m_lst = df_st[['위도','경도']].mean().to_list()

g_map = folium.Map(location = m_lst, zoom_start=15)

for st in st_lst:
    name, long, lat = st[1:]
    
    folium.Marker(location = [lat, long],
                  # icon = folium.Icon(color='orange'),
                  icon = folium.Icon(color='orange', prefix='fa', icon='bus'),
                  popup = '<pre>' + name + '</pre>').add_to(g_map)

g_map
```
