## 네이버 주식 표

```python
import pandas as pd
url = 'https://finance.naver.com/sise/sise_market_sum.nhn?'
tables = pd.read_html(url, encoding='cp949')

df_finance = tables[1]
df_finance = df_finance[df_finance['종목명'].notnull()]
df_finance = df_finance.drop(columns='토론실')
df_finance.columns = ['시가총액순위', '종목명', '현재가','전일비','등락률','액면가','시가총액','상장주식수','외국인비율','거래량', 'PER', 'ROE']
df_finance['시가총액순위'] = df_finance['시가총액순위'].astype('i')
df_finance = df_finance.set_index('시가총액순위')
df_finance
```

## 위키피디아 올림픽 메달 표

```python
import pandas as pd

url = 'https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84'
tables = pd.read_html(url, encoding='utf-8')
tables[0]
```

```python
df_medal = tables[0]
df_medal.columns = ['국가 (IOC 코드)', '하계 참가 횟수', '하계 금', '하계 은', '하계 동',
       '계', '동계 참가 횟수', '동계 금', '동계 은', '동계 동', '계.1',
       '전체 참가 횟수', '전체 금', '전체 은', '전체 동', '총합']
df_medal = df_medal.set_index('국가 (IOC 코드)')
df_medal.head(5)
```

```python
df_medal = df_medal.sort_values('총합', ascending=False)
df_medal.head(5)
```

```python
df_medal.to_csv('올림픽 메달 집계.csv')
df_medal.to_excel('올림픽 메달 집계.xlsx')
```

```python
df_medal2 = df_medal[df_medal['총합'] >= 100]
df_medal2 = df_medal2.drop(index='총합')
df_medal2.head(5)
```

```python
df_medal2['금메달 비중'] = (df_medal2['전체 금'] / df_medal2['총합']).round(2)
df_medal2 = df_medal2.sort_values('금메달 비중', ascending=False)
df_medal2
```
