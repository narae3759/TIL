# 실습 2. 네이버 날씨

```python
from urllib.request import urlopen 
from bs4 import BeautifulSoup

url = 'https://weather.naver.com/'
request = urlopen(url)
html = request.read()
request.status
```

```python
soup = BeautifulSoup(html, 'html.parser')
print(soup.prettify())
```

```python
hourly_weather = soup.find_all('th', class_='data top _cnItemTime')
hourly_weather
```

```python
hourly_weather[0]
```

```python
len(hourly_weather)
```

```python
hourly_weather[0]['data-tmpr']
```

```python
hourly_weather[0]['data-ymdt']
```

```python
hourly_weather[0]['data-wetr-txt']
```

```python
forecast = []
for weather in hourly_weather:
    # print(f"{weather['data-ymdt'][:8]} {weather['data-ymdt'][8:]} {weather['data-tmpr']} {weather['data-wetr-txt']}")
    forecast.append([weather['data-ymdt'][:8], weather['data-ymdt'][8:], int(weather['data-tmpr']), weather['data-wetr-txt']])
forecast
```

```python
import pandas as pd 
df_forecast = pd.DataFrame(forecast)
df_forecast.columns = ['날짜', '시간', '온도', '날씨']
df_forecast.to_csv('네이버시간대별날씨예측.csv')
df_forecast
```

```python
from datetime import datetime
dt = '2022101118'
datetime.strptime(dt, '%Y%m%d%H')
```

```python
forecast2 = []
for weather in hourly_weather:
    dt = datetime.strptime(weather['data-ymdt'], '%Y%m%d%H')
    # print(f"{weather['data-ymdt'][:8]} {weather['data-ymdt'][8:]} {weather['data-tmpr']} {weather['data-wetr-txt']}")
    forecast2.append([dt, int(weather['data-tmpr']), weather['data-wetr-txt']])

df_forecast2 = pd.DataFrame(forecast2)
df_forecast2.columns = ['일시', '온도', '날씨']
df_forecast2 = df_forecast2.set_index('일시')
df_forecast2.to_csv('네이버시간대별날씨예측2.csv')
df_forecast2
```
