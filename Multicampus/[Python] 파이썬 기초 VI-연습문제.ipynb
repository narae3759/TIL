## 정규표현식 연습문제

```python
# match는 처음부터 서식이 맞는 것을 찾는다.
# *는 0개 이상, +는 1개 이상
import re

print(re.match('[0-9]*', '123456'))
print(re.match('[0-9]*', 'abcdef'))
print(re.match('[0-9]+', 'abcdef'))

print(re.match('[a-zA-Z0-9]+', 'Hello1234'))
print(re.match('[a-zA-Z]+', 'Hello1234'))
print(re.match('[0-9]+', 'Hello1234'))

print(re.match('[가-힣]+', '홍길동1234'))
print(re.match('[가-힣0-9]+', '홍길동1234'))
print(re.match('[0-9]+', '홍길동1234'))
```

```python
# ^는 부정. 여집합을 의미 
print(re.match('[^A-Z]+', 'heLLo'))
print(re.match('[^A-Z]+', 'helLo'))
print(re.match('[^A-Z]+', 'HELLO'))
print(re.match('[^A-Z]*', 'heLLo'))
print(re.match('[^A-Z]*', 'helLo'))
print(re.match('[^A-Z]*', 'HELLO'))
```

```python
# search는 중간에도 서식이 맞는 것을 찾는다.
# 찾은 것 중 첫번째 결과만 span으로 나타낸다. 
# 특수문자를 쓰고 싶을 때에는 '\'과 함께 쓴다.

print(re.search('\*+', '1 ** 2'))
print(re.search('[0-9]+', '1 ** 2'))
print(re.search('[$()a-z0-9]+', '$()DoCuMent'))
print(re.search('[$()a-zA-Z0-9]+', '$()DoCuMent'))
```

```python
# \d : [0-9], \D : [^0-9], \w : [a-zA-Z0-9_], \W : [^a-zA-Z0-9_]

print(re.match('\d+', '1234'))
print(re.match('\D+', '12hello34'))
print(re.search('\D+', '12hello34'))
print(re.match('\w+', 'Hello_1234'))
print(re.match('\W+', '$1234'))
```

```python
# 공백 문자 : ' ', '\t', '\n', '\r', '\f', '\v'
# \s : 공백, \S: 공백 제외

print(re.match('[a-zA-Z0-9 ]+', 'Hello 1234'))
print(re.match('[a-zA-Z0-9\s]+', 'Hello 1234'))
print(re.match('[a-zA-Z0-9\s]+', ' Hello 1234'))
print(re.match('[a-zA-Z0-9\S]+', 'Hello\t1234'))
```

```python
# 그룹 묶기
m = re.match('([0-9]+) ([0-9]+)', '10 295')
print(re.match('([0-9]+) ([0-9]+)', '105 29Z66'))
print(m.group(1))
print(m.group(2))
print(m.group())
print(m.group(0))
print(m.groups())       # type tuple
```

```python
# 그룹 이름짓기 (?P<이름> 정규표현식)
m = re.match('(?P<func>[a-zA-Z_][a-zA-Z0-9_]+)\((?P<arg>\w+)\)', 'print(1234)')
print(m)
print(m.group('func'))
print(m.group('arg'))
print(m.group())
print(m.group(0))
print(m.groups())       # type tuple

m = re.match('(?P<func>[a-zA-Z_][a-zA-Z0-9_]+)\((?P<arg>\w*)\)', 'print()')
print(m)
print(m.group('func'))
print(m.group('arg'))
print(m.group())
print(m.group(0))
print(m.groups())       # type tuple
```

```python
def multiple10(m):
    n = int(m.group())
    return str(n * 10)

print(re.sub('[0-9]+', multiple10, '1 2 Fizz 4 Buzz Fizz 7 8'))
print(re.sub('[0-9]+', lambda m: str(int(m.group())*10), '1 2 Fizz 4 Buzz Fizz 7 8'))
```

```python
# 연습문제 
# 왜 유효하지 않은지 출력하는 기능 만들기 
# 비밀번호를 등록하시겠습니까? 문구 만들기
# 조건1. 대문자와 소문자와 숫자 모두 포함해야 한다. 
# 조건2. 특수문자를 포함해야 한다.
 
import re 

while True:
    text = input('패스워드를 입력해주세요. 종료하고 싶으시면 999를 눌러주세요')
    if text == '999':
        break
    print('입력한 패스워드:', text)
    if len(text) < 8:
        print('8자리 이상 입력하셔야 합니다. ')
        continue
    else:
        match = re.search('[a-zA-Z0-9_@$!]*[a-zA-Z0-9]+[_@$!]+[a-zA-Z0-9_@$!]*',text)
        if match == None:
            print('유효하지 않은 패스워드')
            continue
        print('유효한 패스워드입니다!')
        break
```

```python
# 연습문제
Str = '''
눈 길 닿는 곳마다, 거리 풍경도 가을옷으로 갈아입고 있는데요.
오늘 설악산 중청대피소에서 찍은 사진입니다.
곳곳에 붉게 물들어가는 나무들이 보이실텐데, 보통, 산 정상부의 약 20%가 물들면, '첫 단풍'이라고 말합니다.
이번주 목요일 경이면, 설악산은 첫 단풍이 될 것으로 예상을 하고 있습니다.
이처럼 가을은 깊어가고 있고요.
내일도 일교차는 10-15도 안팎까지 크게 나겠습니다.
아침에 서울 17도, 강원도 대관령 6도, 경북 안동 12도로 쌀쌀하겠고요.
낮에는 서울 27도까지 오르는 등, 평년수준을 3도가량 웃돌겠습니다.
위성 영상 보시면, 현재 전국적으로 맑은 하늘이 드러나 있는데요.
동풍의 영향으로 동해안은 밤한때 빗방울이 살짝 떨어지겠고요.
내일 새벽부터 아침사이 내륙에는 짙은 안개가 끼겠습니다.
'''

def Encript(Str, shift=100):
    return chr(ord(Str)+shift)

def Decode(Str, shift=100):
    return chr(ord(Str)+shift)

shift = 100
print(list(map(Encript, Str, [shift]*len(Str))))
Str_e = list(map(Encript, Str, [shift]*len(Str)))

shift = -100
print(list(map(Decode, Str_e, [shift]*len(Str))))
Str_d = list(map(Decode, Str_e, [shift]*len(Str)))
print(''.join(Str_d))
```

```python
# 연습문제 
# text = input('암호화할 텍스트를 입력하세요').upper()
text = 'ATTACK TO MIDNIGHT'
shift = 3

def Caesar(text, shift):
    src_str = string.ascii_uppercase
    dst_str = src_str[shift:] + src_str[:shift]

    encript_text = ''
    for char in text:
        if char == ' ':
            encript_text += ' '
        else:
            ind = src_str.index(char)
            encript_text += dst_str[ind]

    return encript_text
    
def Decoding(text, shift):
    src_str = string.ascii_uppercase
    dst_str = src_str[shift:] + src_str[:shift]

    decoding_text = ''
    for char in text:
        if char == ' ':
            decoding_text += ' '
        else:
            ind = dst_str.index(char)
            decoding_text += src_str[ind]

    return decoding_text
    
encript_text = Caesar(text,shift)

print(encript_text)
print(Decoding(encript_text, shift=3))
```

# 워드 클라우드

```python
import wikipedia
from wordcloud import WordCloud, STOPWORDS

wiki = wikipedia.page('Arttificial intelligence')
text = wiki.content

print('단어 수:', len(text.split()))

wordcloud = WordCloud(width = 2000, height = 1500).generate(text)

import matplotlib.pyplot as plt

plt.figure(figsize = (10, 10))
plt.imshow(wordcloud)
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS

s_words = STOPWORDS.union({'one','using','first','two','make','use'})
wordcloud = WordCloud(width = 2000, height = 1500, stopwords = s_words).generate(text)

import matplotlib.pyplot as plt

plt.figure(figsize = (10, 10))
plt.imshow(wordcloud)
plt.show()
```
