```python
# from urllib.request import urlopen 
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
def get_details(items):
    title = item.find('span', class_='text__item').text.strip()
    try:
        brand = item.find('span', class_='text__brand').text
    except:
        brand = ""
    try:
        price = item.find('strong', class_='text text__value').text
        price = int(price.strip().replace(',',''))
    except:
        price = ""
    try:
        rate_cnt = item.find('li', class_='list-item list-item__feedback-count').text
        rate_cnt = re.sub('[^0-9]','',rate_cnt)
    except:
        rate_cnt = ""
    try:
        rating = item.find('span', class_='image__awards-points')['style']
        rating = re.sub('[^0-9]','',rating)
    except:
        rating = ""
    try:
        sold_cnt = item.find('li', class_='list-item list-item__pay-count').text
        sold_cnt = re.sub('[^0-9]','',sold_cnt)
    except:
        sold_cnt = ""
    
    return [title, brand, price, rate_cnt, rating, sold_cnt]
```

```python
keyword = input('조회할 제품을 입력하세요: ')
no_of_pages = int(input('페이지 수를 입력하세요.'))
option = 13         # 상품평 많은 순

item_list = []
for page in range(1,no_of_pages+1):
    url = f'https://browse.gmarket.co.kr/search?keyword={keyword}&s={option}&k=0&p={page}'
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    
    items = soup.find_all('div', {'class':'box__item-container'})
    
    for item in items:
        item_detail = get_details(item)
        item_list.append(item_detail)

df_gmarket = pd.DataFrame(item_list)
df_gmarket.columns = ['제품명', '브랜드', '가격', '상품평', '평점', '구매 수']
df_gmarket.to_csv('Gmarket-{}.csv'.format(keyword))

print('저장이 완료되었습니다.')
df_gmarket
```

```python
import os
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import re
from datetime import datetime
```

```python
def item_detail(item):
    title = item.find('a', class_='tit_main').text
    
    link = 'https://search.daum.net/search' + item.find('a', class_='thumb')['href']
    image = item.find('img', class_='thumb_img')['src']
    img_data = requests.get(image)
    with open(f"Movies_img/{year}_{re.sub('[^0-9a-zA-Z가-힣._]','',title)}.png", 'wb') as f:
        f.write(img_data.content)
    # print(f'=========={title} 이미지 파일을 저장했습니다.==========')
    
    rate = float(item.find('em', class_='rate').text)
    reviewer = item.find('a', class_='link_count').text
    
    dls = list(item.find('dl', class_='dl_comm').next_siblings)
    dls = [dl for dl in dls if dl != ' ']
    
    open_type = dls[0].find('dt', class_='tit_base').text.strip()
    open_date = dls[0].find('dd', class_='cont').text.strip()
    open_date = datetime.strptime(open_date, '%Y.%m.%d.')
    year_views = dls[1].find('dd', class_='cont').text
    cum_views = dls[2].find('dd', class_='cont').text
    
    only_num = [reviewer, year_views, cum_views]
    for ind, lst in enumerate(only_num):
        only_num[ind] = int(re.sub('[^0-9]', '', lst))
        
    return [title, rate, only_num[0], open_type, open_date, only_num[1], only_num[2], link]
```

```python
file_path = 'Movies_img/'
if not os.path.isdir(file_path):
    os.mkdir(file_path)

start = 2020
end = 2021

item_lst = []

for year in range(end, start-1, -1):
    url = f'https://search.daum.net/search?w=tot&q={year}%EB%85%84%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84&DA=MOR&rtmaxcoll=MOR'
    response = requests.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    temp = soup.find('ol', class_='type_plural list_exact movie_list').li
    items = [item for item in list(temp.next_siblings) if item != ' ']
    items.insert(0, temp)
    
    for item in items:
        details = item_detail(item)
        details.insert(0, year)
        item_lst.append(details)

df_movie = pd.DataFrame(item_lst)
df_movie.columns = ['년도','영화 제목', '평점', '평점 참여자', '구분', '(재)개봉일', '연간 관객 수', '누적 관객 수', '링크']
df_movie.to_csv('다음영화스크래핑.csv')
df_movie
```

# 실습 7. 네이버 TV

```python
# (p169) 기사별 데이터 수집 
import requests
from bs4 import BeautifulSoup

url = "https://tv.naver.com/r/"
headers = {'User-Agent':'Mozilla/5.0'}
response = requests.get(url, headers=headers)
html = response.text
soup = BeautifulSoup(html, "html.parser")


videos = soup.select("dl.cds_info")
videos
```

```python
def video_detail(video):
    title = video.find('dt', class_='title').text.strip('\n')
    channel = video.find('dd').text.strip('\n')
    plays = video.find('span', class_='hit').text
    plays = int(re.sub('[^0-9]','', plays))
    likes = video.find('span', class_='like').text
    likes = int(re.sub('[^0-9]','', likes))

    return [title, channel, plays, likes]

```

```python
videos[4].find('div', class_='info')
```

```python
i = 0
video_lst = []

for video in videos:
    i += 1
    temp = [i]
    temp.extend(video_detail(video))
    video_lst.append(temp)
    
video_lst
```

```python
df_naver_video = pd.DataFrame(video_lst)
df_naver_video.columns = ['순위', '제목', '채널', '재생 수', '좋아요 수']
df_naver_video.to_excel('네이버tv_top100.xlsx')
df_naver_video
```
